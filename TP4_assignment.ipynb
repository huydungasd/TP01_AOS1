First, we download the image for the dataset at: http://lesun.weebly.com/hyperspectral-data-set.html . The image uesed in this TP is PaviaU.

```python
import scipy.io as sio
import matplotlib.pyplot as plt
import numpy as np

image_mat = sio.loadmat('PaviaU.mat')['paviaU']
# Reshape the 3D dataset to 2D dataset
image_mat_lin = image_mat.reshape((-1, image_mat.shape[2]))
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import MinMaxScaler

# Initialize an instance of PCA's class with 3 principle components and fit it
pca = PCA(n_components=3)
Data_new = pca.fit_transform(image_mat_lin)

# Rescale the data to [0,1] to be displayed by using plt.imshow
Data_new = MinMaxScaler().fit_transform(Data_new)

# Reshape the data to M*N*3
image_mat_new = Data_new.reshape((image_mat.shape[0], image_mat.shape[1], 3))
plt.imshow(image_mat_new)
```
